```python
import matplotlib.pyplot as plt
```

```python
# Num/Num
x=[23,45,67,12]
y=[2,4,6,8]
plt.plot(x,y)
```

```python
# Num/Category
x=[23,45,67,12]
y=[2,4,6,8]
z=["A","B","C","D"]
plt.plot(x,z)
```

```python
# Num/Category
x=[23,45,67,12]
y=[2,4,6,8]
z=["A","B","C","D"]
plt.plot(x,y,x,z)
```

```python
# Num/Category
x=[23,45,67,12]
y=[2,4,6,8]
z=["A","B","C","D"]
plt.plot(x,y,x,y)
```

```python
# Category/Category
# x=[23,45,67,12]
# y=[2,4,6,8]
z=["A","B","C","D"]
p=[True,False,False,True]
plt.plot(z,p)
```

```python
# Category/Category
# x=[23,45,67,12]
# y=[2,4,6,8]
# z=["A","B","C","D"]
p=[True,False,False,True]
q=[False,True,False,True]
plt.plot(p,q)
```

```python
p=[True,False,False,True]
q=[False,True,False,True]
plt.plot(p,q)
```

# barchart

```python
x=[89,78,90,95,104]
y=[105,18,132,20,23]
plt.bar(x,y)
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
x=["X","y","z","W","r"]
y=[105,18,132,20,23]
plt.bar(x,y)
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
x=["X","y","z","W","r"]
y=["a","n","k","u","e"]
plt.bar(x,y)
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
x=["X","y","z","W","r"]
y=[105,18,132,20,23]
plt.bar(x,y,color="c")
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
x=["X","y","z","W","r"]
y=[105,18,132,20,23]
plt.barh(x,y,)
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.figure(figsize=(12,7))
x=["X","y","z","W","r"]
y=[105,18,132,20,23]
p=["p","q","q","w","o"]
q=[23,45,67,54,32]
plt.bar(x,y,width=0.3)
plt.bar(p,q)
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

# piechart

```python
a=[23,34,56,78]
ml=["a","b","c","d"]
plt.pie(a,labels=ml)
```

```python
a=[23,34,56,78]
ml=["a","b","c","d"]
plt.pie(a,labels=ml,autopct="%0.3f%%")
```
